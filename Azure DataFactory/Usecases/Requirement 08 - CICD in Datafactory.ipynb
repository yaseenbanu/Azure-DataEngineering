# Manage Tab

Under Manage like Git Configuration and then select a collaboration branch(Each time you save it this will get stored) and Publish branch(Config changes) in github.

Depending on the repository you picked, you have the option  to switch between multiple branches. You can even create and do a pull request from the options.

# Understanding Branches

- Collaboration Branch: Contains your development and pipeline changes.
- Main Branch: Where stable and tested changes are merged.
- ADF_publish Branch: Automatically updated during the publish process. It includes the configuration and state of the production environment based on the Main Branch.


# Workflow:

1.	Develop: Make and test pipeline changes in the Collaboration Branch.
2.	Merge: Merge these changes into the Main Branch.
3.	Publish: Publish from the Main Branch to update the ADF_publish Branch with configuration changes that will be deployed to production.
